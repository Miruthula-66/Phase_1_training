```
file_location = "/FileStore/tables/matches.csv"
df = spark.read.option("header", True).option("inferSchema", True).csv(file_location)
df.printSchema()
df.show(5)
```

```
teams = df.select("team1").union(df.select("team2")).distinct()
teams_count = teams.count()
print(f"Number of unique teams: {teams_count}")
teams.show(truncate=False)
df.select("result").distinct().show()
completed_matches = df.filter(df.result != "no result")
completed_matches.count()
cleaned_df = completed_matches.drop("umpire3", "dl_applied")
cleaned_df.printSchema()
```
